```python
# Importando bibliotecas
import requests
import pandas as pd
```

```python
# Criando uma lista de IDs
list_ids = [ids for ids in range(1, 100)]
```

```python
# Verificando se a API está funcionando
url = f'https://rickandmortyapi.com/api/character/{",".join(map(str, list_ids))}'

try:
    response = requests.get(url)

    # Verifique se a resposta tem um código de status 200, indicando sucesso
    response.raise_for_status()

    print(f'A API está funcionando corretamente. Código de status: {response.status_code}')
except requests.exceptions.RequestException as e:
    print(f'Erro ao acessar a API: {e}')

# Criando uma variável que irá receber os dados da API no formato JSON
data = requests.get(url).json()
```
